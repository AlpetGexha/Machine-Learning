```python
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv("dataset2.csv")
data.head()
```

```python
fig = plt.figure()
# grapth margins and size xmin, ymin, width, and height
ax1 = fig.add_axes([0.1,0.1,0.8,0.8])
ax1.plot(data['Close'])
ax1.set_xlabel("Date")
ax1.set_ylabel("Prices")
ax1.set_title("Close prices data")
plt.show()
```

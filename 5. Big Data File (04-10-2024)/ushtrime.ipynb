```python
def feet_inches_to_cm(height):
    if isinstance(height, str):
        # Split the height string on the '-'
        feet, inches = height.split('-')
        # Convert feet to cm and inches to cm, then sum them
        cm = (int(feet) * 30.48) + (int(inches) * 2.54)
        return cm
    return None


def bmi(weight, height):
    return weight / (height / 100) ** 2


height_cm = data['Height'].apply(feet_inches_to_cm)
data['BMI'].apply(bmi, height=height_cm)

data.head()
```

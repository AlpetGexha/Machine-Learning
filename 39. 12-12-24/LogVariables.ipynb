SI te transformojme te dhenat e y-variable ne Python (Log, Square Root, Cube Root)

Kur ne hasim problemi e heteroskedasticitetit duhet te rregullojme variablin e varur ne menyre qe te eliminojme kete fenomen te shperndarjes se te dhenave

1. Log Transformation: Transformon y ne log(y).

2. Square Root Transformation:  Transformon y ne âˆšy.

3. Cube Root Transformation: Transformon y ne y1/3.

Duke bere nje nga keto 3 transformime te dhenat kane nje shperndarje me normale

```python
import numpy as np
import matplotlib.pyplot as plt
```

Log Transformation in Python

```python
np.random.seed(0)
data_y = np.random.beta(a = 4, b =15, size = 300)
```

```python
data_y_log = np.log(data_y)

fig, axs = plt.subplots(nrows=1, ncols=2)
axs[0].hist(data_y)
axs[1].hist(data_y_log)

axs[0].set_title('Original Data')
axs[1].set_title('Log_Transformed data')
```

Square Root Transformation in Python

```python

np.random.seed(0)
data_y = np.random.beta(a = 4, b =15, size = 300)

data_y_sq = np.sqrt(data_y)

fig, axs = plt.subplots(nrows=1, ncols=2)
axs[0].hist(data_y)
axs[1].hist(data_y_sq)

axs[0].set_title('Original Data')
axs[1].set_title('Square root Transformed data')
```

Cube Root Transformation in Python

```python

np.random.seed(0)
data_y = np.random.beta(a = 4, b =15, size = 300)

data_y_cb = np.cbrt(data_y)

fig, axs = plt.subplots(nrows=1, ncols=2)
axs[0].hist(data_y)
axs[1].hist(data_y_cb)

axs[0].set_title('Original Data')
axs[1].set_title('Cube root Transformed data')
```

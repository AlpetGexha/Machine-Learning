```python

import matplotlib.pyplot as plt
import numpy as np
```

```python
# line graph midis dy te dhenave
boshti_x = np.array([0, 5])
boshti_y = np.array([0, 55])
plt.plot(boshti_x, boshti_y)
```

```python
# grafik me dy pika identifikuese
boshti_x = np.array([0, 5])
boshti_y = np.array([0, 55])
plt.plot(boshti_x, boshti_y, 'o')
```

```python
# grafik line ku jan ete evidentuara dhe pikat e dhena
boshti_x = np.array([2, 3, 4, 5, 7, 9, 10])
boshti_y = np.array([1, 4, 6, 7, 3, 7, 10])
plt.plot(boshti_x, boshti_y, marker='*')
```

```python
# te njejtin grafik si me siper ndryshojme ngjyra-vija...
boshti_x = np.array([2, 3, 4, 5, 7, 9, 10])
boshti_y = np.array([1, 4, 6, 7, 3, 7, 10])
plt.plot(boshti_x, boshti_y, '*--g', ms=6)
```

```python
boshti_x = np.array([2, 3, 4, 5, 7, 9, 10])
boshti_y = np.array([1, 4, 6, 7, 3, 7, 10])
# linestyle = 'dashed'
plt.plot(boshti_x, boshti_y, ls='--', color='y', lw='8')
```

```python
# grafiku me dy linja
data_1 = np.array([2, 4, 6, 9, 10])
data_2 = np.array([5, 9, 10, 4, 7])
plt.plot(data_1)
plt.plot(data_2)
```

```python
# bejme nje grafik me dy linja-dy te dhena
data1_x = np.array([2000, 2005, 2010, 2015, 2020, 2024])
data1_y = np.array([23, 26, 22, 27, 20, 29])

data2_x = np.array([2002, 2007, 2012, 2017, 2022, 2024])
data2_y = np.array([18, 24, 22, 24, 21, 25])

# plt.plot(data1_x, data1_y, data2_x, data2_y, marker = 'o')
plt.plot(data1_x, data1_y, marker='o', color='r')
plt.plot(data2_x, data2_y, marker='o', color='k')

plt.title("Temperatura mesatare ne periudha 5 vjecare")
plt.xlabel('Vitet')
plt.ylabel('Temperatura ne grade')
```

```python
# Grafik me legjende
# plt.plot(data1_x, data1_y, data2_x, data2_y, marker = 'o')
plt.plot(data1_x, data1_y, marker='o', color='r', label='Temp ne Shqiperi')
plt.plot(data2_x, data2_y, marker='o', color='k', label='Temp ne Kosove')

plt.title("Temperatura mesatare ne periudha 5 vjecare")
plt.xlabel('Vitet')
plt.ylabel('Temperatura ne grade')
plt.legend()
```

```python
# subplot
data1_x = np.array([2000, 2005, 2010, 2015, 2020, 2024])
data1_y = np.array([23, 26, 22, 27, 20, 29])
plt.subplot(1, 2, 1)
plt.plot(data1_x, data1_y, marker='o', color='r')
plt.title('Temperatura ne Shqiperi', fontsize=8)
plt.xlabel('Vitet', fontsize=8)
plt.ylabel('Temperatura ne grade', fontsize=8)
plt.grid()

data2_x = np.array([2002, 2007, 2012, 2017, 2022, 2024])
data2_y = np.array([18, 24, 22, 24, 21, 25])
plt.subplot(1, 2, 2)
plt.plot(data2_x, data2_y, marker='o', color='k')
plt.title('Temperatura ne Kosove', fontsize=8)
plt.xlabel('Vitet', fontsize=8)
plt.grid()

plt.suptitle("Temperatura mesatare ne periudha 5 vjecare")
```

```python
# scatter plot
data1 = np.array([2, 3, 4, 6, 8, 9, 10, 14, 16, 19, 20, 25])
data2 = np.array([7, 9, 15, 16, 17, 22, 24, 56, 27, 30, 32, 33])

plt.scatter(data1, data2, label='Pagesa per sasi te marre', color='g')
plt.legend()
```

```python
# barchart
data_1 = ['shume mire', 'mire', 'mjaftueshem', 'keq']
data_2 = [34, 56, 56, 76]
plt.bar(data_1, data_2)
plt.xlabel('Vleresimi')
plt.ylabel('Frekuenca')
plt.title('Vleresimi mbi projektin')
```

```python
# piechart
vlerat = [10, 20, 35, 15]
label_vlera = ['molle', 'portokalle', 'dardhe', 'mandarin']

plt.pie(vlerat, labels=label_vlera, autopct='%1.1f%%')
```

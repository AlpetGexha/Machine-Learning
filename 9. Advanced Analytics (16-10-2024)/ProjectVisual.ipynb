```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
# importimin e datasetit
dataseti = pd.read_csv('dataseti.csv')
```

```python
dataseti.info()
```

```python
dataseti.head()
```

```python
dataseti.drop_duplicates()
```

```python
dataseti.info()
```

```python
# kthej kolone timestamp ne nje datetime
from datetime import datetime

dataseti['timestamp']=pd.to_datetime(dataseti['timestamp'])
```

```python
dataseti.info()
```

```python
dataseti['year'] = pd.DatetimeIndex(dataseti['timestamp']).year
```

```python
dataseti.head()
```

```python
dataseti.info()
```

```python
dataseti['item_id'].nunique() # kemi 9560 lloje produktesh te shitura
```

```python
dataseti['user_id'].nunique() # nr i klieneteve
```

```python
dataseti['user_id']= dataseti['user_id'].astype(str)
dataseti['item_id']= dataseti['item_id'].astype(str)
```

```python
dataseti.info()
```

```python
dataseti.describe()
```

```python
# Vizualizime
# grafiku i frekuences se rating

sb.countplot(x='rating', data=dataseti)
```

```python
# bejme vizualizimin e viteve
dataseti.groupby('year')['rating'].count().plot(kind='bar')
```

```python
new_dataset = dataseti[((dataseti['year']==2014) | (dataseti['year']==2015) | (dataseti['year']==2016) | (dataseti['year']==2017))]
```

```python
new_dataset.info()
```

```python
# bej vizualizimin e category
new_dataset.groupby('category')['rating'].count().plot(kind='bar')
```

```python
# ndertoj grafik per brand product
new_dataset.groupby('brand')['rating'].count().plot(kind='bar')
```

```python
# grafikun me siper e bejme me te qarte, duke marre me pak brand, me te shiturat
new_dataset.groupby('brand')['rating'].count().sort_values(ascending=False).head(14).plot(kind='bar')
```

```python
# grafikun me siper e bejme me te qarte, duke marre me pak brand, me pak te shiturat
new_dataset.groupby('brand')['rating'].count().sort_values().head(10).plot(kind='bar')
```

```python
# paraqes grafikun e muajve
new_dataset['month'] = pd.DatetimeIndex(new_dataset['timestamp']).month
new_dataset.head()
```

```python
new_dataset.groupby('month')['rating'].count().plot(kind='bar')
```

```python
new_dataset.groupby('brand')['rating'].count().sort_values(ascending=False).head(5).plot(kind='pie', autopct='%1.1f%%', shadow=True)
```

```python
new_dataset.groupby('category')['rating'].count().sort_values(ascending=True).head().plot(kind='pie', autopct='%1.1f%%', shadow=True)
```

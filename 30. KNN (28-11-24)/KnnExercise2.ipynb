```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
import seaborn  as sns
```

```python
X, y = make_blobs(n_samples = 500, n_features = 2, centers = 4,cluster_std = 1.5, random_state = 14)

X
```

```python
y
```

```python
plt.figure(figsize=(10,10))
plt.scatter(X[:, 0], X[:,1], c=y, marker='*', s=100)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=23)
```

```python
knn1 = KNeighborsClassifier(n_neighbors=1)
knn5 = KNeighborsClassifier(n_neighbors=5)
```

```python
knn1.fit(X_train, y_train)
knn5.fit(X_train, y_train)
```

```python
y_pred_knn1 = knn1.predict(X_test)
y_pred_knn5 = knn5.predict(X_test)
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.scatter(X_test[:, 0], X_test[:,1], c=y_pred_knn1, marker='*', s=100)
plt.title('Vlerat e parashikuara per neighbors= 1')
plt.subplot(1,2,2)
plt.scatter(X_test[:, 0], X_test[:,1], c=y_pred_knn5, marker='*', s=100)
plt.title('Vlerat e parashikuara per neighbors= 5')
```

```python
from sklearn.metrics import accuracy_score
acc_sc_knn1 = accuracy_score(y_test, y_pred_knn1)
acc_sc_knn1
```

```python
acc_sc_knn5 = accuracy_score(y_test, y_pred_knn5)
acc_sc_knn5
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, y_pred_knn1))
print(classification_report(y_test, y_pred_knn1))
```

```python
print(confusion_matrix(y_test, y_pred_knn5))
print(classification_report(y_test, y_pred_knn5))
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
sns.heatmap(confusion_matrix(y_test, y_pred_knn1), annot=True, cmap='YlGnBu')
plt.title('Matrica neighbors= 1')
plt.subplot(1,2,2)
sns.heatmap(confusion_matrix(y_test, y_pred_knn5), annot=True, cmap='YlGnBu')
plt.title('Matrica neighbors= 5')
```
